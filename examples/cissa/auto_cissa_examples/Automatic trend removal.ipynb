# Automatic Trend Removal

```python
'''This notebook contains an example of automatically extracting the trend component of a signal using Cissa. Useful if you need an easy way to detrend a time series.'''
```

```python
#importing required packages
import pandas as pd
import os
dir_path = os.path.abspath('')
os.chdir('../../..')
home_path = os.path.abspath('')
os.chdir('./pycissa/processing/cissa')
from cissa import Cissa
os.chdir(home_path)
```

```python
## Read data file  (source is the Ministry of Economy of Spain.)
data = pd.read_excel(home_path + r'/data/EPC.xlsx',sheet_name = 'Data')
```

```python
#create the cissa object
cissa_object = Cissa(t=data['Date'],x=data['Electric power consumption in Spain'])
```

```python
#run detrending.We choose L = 96 (8 years)
#this makes the definition of trend and period longer than 8 years
cissa_object.auto_detrend(L = 96,t_unit='month',data_unit='Mega-Watts',timestep=60*60*24*30)
```

## Plot the trend and detrended signal

```python
#Plot the trend and detrended signal
cissa_object.figures.get('cissa').get('figure_detrended')
```

## Analyse the trend

```python
cissa_object.figures.get('cissa').get('figure_trend')
```

```python
cissa_object.results.get('cissa').get('trend results').get('rolling_OLS').get('trend_slope')
```

## We can access the trend and detrended signal as numpy arrays for further use:

```python
#Trend
cissa_object.x_trend
```

```python
#Detrended:
cissa_object.x_detrended
```

```python
'''We can try again with a shorter window length L'''
#run detrending.We choose L = 24 (2 years)
cissa_object.auto_detrend(L = 24)
```

```python
#Plot the trend and detrended signal for L = 24 (2 years)
#this makes the definition of trend and period longer than 2 years
cissa_object.figures.get('cissa').get('figure_detrended')
```
